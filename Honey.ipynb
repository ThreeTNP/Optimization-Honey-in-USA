```python
from sklearn.linear_model import LinearRegression
from sklearn.utils.fixes import parse_version, sp_version

# This is line is to avoid incompatibility if older SciPy version.
# You should use `solver="highs"` with recent version of SciPy.
solver = "highs" if sp_version >= parse_version("1.6.0") else "interior-point"
```

```python
from sklearn.linear_model import QuantileRegressor
```

### Production From Each State

```python
j = 1
predict_honey = {}
score_quantile = {}
for i in data.state.unique():
  lr = LinearRegression()
  q1 = QuantileRegressor(quantile=0.05, solver=solver, alpha=0)
  q2 = QuantileRegressor(quantile=0.5, solver=solver, alpha=0)
  q3 = QuantileRegressor(quantile=0.95, solver=solver, alpha=0)
  y = data[data["state"]==i].loc[:, ["production"]]
  X = data[data["state"]==i].loc[:, ["year"]]
  q1.fit(X, y)
  q2.fit(X, y)
  q3.fit(X, y)
  lr.fit(X, y)
  predict_honey[i] = lr.predict([[2024]])
  plt.scatter(X, y, alpha=0.2)
  plt.plot(X, y, '-o')
  plt.plot(X, lr.predict(X), label="Linear Regression", linestyle='--')
  plt.plot(X, q1.predict(X), label="05th Percentile")
  plt.plot(X, q2.predict(X), label="50th Percentile")
  plt.plot(X, q3.predict(X), label="95th Percentile")
  plt.title(i)
  plt.xlabel("Year")
  plt.ylabel("Production")
  plt.legend()
  plt.show()
  j += 1
```

### AVG Price From Each State

```python
predict_price = {}
for i in data.state.unique():
  lr = LinearRegression()
  q1 = QuantileRegressor(quantile=0.05, solver=solver, alpha=0)
  q2 = QuantileRegressor(quantile=0.5, solver=solver, alpha=0)
  q3 = QuantileRegressor(quantile=0.95, solver=solver, alpha=0)
  y = np.array(data.average_price[data["state"]==i]).reshape(-1, 1)
  X = np.array(data.year[data["state"]==i]).reshape(-1, 1)
  lr.fit(X, y)
  q1.fit(X, y)
  q2.fit(X, y)
  q3.fit(X, y)
  predict_price[i] = lr.predict([[2024]])
  plt.scatter(X, y)
  plt.plot(X, y, '-o')
  plt.plot(X, lr.predict(X), label="Linear Regression", linestyle='--')
  plt.plot(X, q1.predict(X), label="05th Percentile")
  plt.plot(X, q2.predict(X), label="50th Percentile")
  plt.plot(X, q3.predict(X), label="95th Percentile")
  plt.xlabel("Year")
  plt.ylabel("Average Price")
  plt.title(i)
  plt.legend()
  plt.show()
```

```python
quantile_range = np.arange(0.05, 1, 0.05)
optimize_quan_avgprice = {}
optimize_avgprice = {}
score_avgprice = {}
for i in data.state.unique():
  keep_quan = {}
  y = np.array(data.average_price[data["state"]==i]).reshape(-1, 1)
  X = np.array(data.year[data["state"]==i]).reshape(-1, 1)
  for quan in quantile_range:
    q = QuantileRegressor(quantile=quan, solver=solver, alpha=0)
    q.fit(X, y)
    keep_quan[quan] = q.score(X, y)
  opt = [i for i in keep_quan if keep_quan[i] == sorted(keep_quan.values(), reverse=True)[0]]
  optimize_quan_avgprice[i] = opt[0]
  q2 = QuantileRegressor(quantile=optimize_quan_avgprice[i], solver=solver, alpha=0)
  q2.fit(X, y)
  score_avgprice[i] = q2.score(X, y)
  optimize_avgprice[i] = q2.predict([[2024]])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import QuantileRegressor
from sklearn.utils.fixes import parse_version, sp_version
from sklearn.metrics import mean_absolute_error, mean_squared_error

solver = "highs" if sp_version >= parse_version("1.6.0") else "interior-point"

for i in data.state.unique():
  X = data[data["state"]==i].loc[:, ["year"]]
  y = data[data["state"]==i].loc[:, ["production"]]

  linear_regression = LinearRegression()
  quantile_regression = QuantileRegressor(quantile=0.5, alpha=0, solver=solver)

  y_pred_lr = linear_regression.fit(X, y).predict(X)
  y_pred_qr = quantile_regression.fit(X, y).predict(X)

  print(f"""Year Production error in {i}
    {linear_regression.__class__.__name__}:
    MAE = {mean_absolute_error(y, y_pred_lr):.3f}
    MSE = {mean_squared_error(y, y_pred_lr):.3f}
    {quantile_regression.__class__.__name__}:
    MAE = {mean_absolute_error(y, y_pred_qr):.3f}
    MSE = {mean_squared_error(y, y_pred_qr):.3f}
    """)
```
